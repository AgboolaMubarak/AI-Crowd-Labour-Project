```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
train=pd.read_csv("train (4).csv")
test=pd.read_csv("test (3).csv")
```

```python
train.info()
```

```python
test.info()
```

```python
train.head()
```

```python
train["total wageincrease"]=train["wage-increase-first-year"] + train["wage-increase-second-year"] + train["wage-increase-third-year"]
```

```python
test["total wageincrease"]=test["wage-increase-first-year"] + test["wage-increase-second-year"] + test["wage-increase-third-year"]
```

```python
X=train.drop(["class"],axis=1)
y=train["class"]
```

```python
from catboost import CatBoostClassifier
```

```python
model=CatBoostClassifier(iterations=2000,learning_rate=0.1,depth=7)
```

```python
model.fit(X,y)
```

```python
predictions=model.predict(test)
```

```python
sub=pd.DataFrame()
sub["class"]=predictions
```

```python
sub.to_csv("labor6.csv",index=False)
```

```python
import lightgbm as lgb
```

```python
clf = lgb.LGBMClassifier(num_leaves=31,
    max_depth=-1,
    learning_rate=0.1,
    n_estimators=2000)
```

```python
clf.fit(X,y)
```

```python
pred=clf.predict(test)
```

```python
sub2=pd.DataFrame()
sub2["class"]=pred
```

```python
sub2.to_csv("labor2.csv",index=False)
```

```python
from xgboost import XGBClassifier
```

```python
xgb=XGBClassifier(max_depth=7,learning_rate=0.1,n_estimators=2000)
```

```python
xgb.fit(X,y)
```

```python
xgbpred=xgb.predict(test)
```

```python
sub3=pd.DataFrame()
sub3["class"]=xgbpred
```

```python
sub3.to_csv("labor3.csv",index=False)
```

```python
ultimatesub=pd.DataFrame()
ultimatesub["class"]=(0.5 * sub["class"])+ (0.2 * sub2["class"]) + (0.3 * sub3["class"])
ultimatesub["class"]=ultimatesub["class"].astype(int)
ultimatesub.to_csv("labor5.csv",index=False)
```
